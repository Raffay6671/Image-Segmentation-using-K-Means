```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

image = Image.open("E:\\check.jpeg")  # Make sure to use double backslashes or a raw string

image_array = np.array(image)

height, width, channels = image_array.shape

pixels = image_array.reshape(-1, channels)

K = 3

initial_centroids_indices = np.random.choice(len(pixels), size=K, replace=False)
centroids = pixels[initial_centroids_indices]

max_iterations = 100

for _ in range(max_iterations):
    distances = np.linalg.norm(pixels[:, np.newaxis] - centroids, axis=2)

    labels = np.argmin(distances, axis=1)

    for i in range(K):
        cluster_points = pixels[labels == i]
        if len(cluster_points) > 0:
            centroids[i] = np.mean(cluster_points, axis=0)

segmented_image = centroids[labels].reshape((height, width, channels)).astype(np.uint8)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(segmented_image)
plt.title("Segmented Image")

plt.show()
```
